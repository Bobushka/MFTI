## Лекция 7: Математическая статистика - Матматическое ожидание и дисперсия <a name="Лекция7:Математическаястатистика-Матматическоеожиданиеидисперсия"></a>

### Свойства математического ожидания
- свойство линейности:  
 - М(a\*x + b) = a\*M(x) + b, где а и b - константы  
 - M(x1 + x2 + ...) = M(x1) + M(x2) + ...
- свойство дистрибутивности:  
 - M(x1 * x2 * ...) = M(x1) * M(x2) * ..., но только если х1, х2, ... - н.с.в.

#### Дисперсия и ее свойства:  
Напомним, что Дисперсия есть центральный момент второго порядка от математического ожидания.  
Свойства дисперсии: 
 - D(x) = M(x\*\*2) - (M(x))\*\*2  
 - D(ax + b) = a\*\*2D(x)  

#### Ковариация и ее свойства:
Коэффициент ковариации (cov) двух величин x и  y: 
 - определение: cov(x, y) = M (x - M(x)) (y - M(y))
 - свойство_1: cov(x, y) = M(x\*y) - M(x)\*M(y)
 - свойство_2: cov(x, x) = D(x)

#### Еще свойства:
- Дисперсия суммы равна сумме дисперсий плюс сумма попарных ковариаций  
- Если х и y - н.с.в., то cov(x, y) = 0. Обратное неверно.
- Если х1, х2, ... - независимы попарно, то Дисперсия суммы равна сумме дисперсий
- Свойство неотрицательной определенности матрицы

### Неравенства, связанные с математическим ожиданием
- Мат ожидание от неотрицательной величины тоже неотрицательно
- Если x меньше y, то и М(х) меньше M(y)
- Неравенство Чебышёва: вероятность отклонения величины о ее математического ожидания пропорциональна ее дисперсии  
P(|x - M(x)| > e) <= D/e\*\*2
- Если дисперсия равна нулю, то величина совпадает со своем мат ожиданием. То есть, является постоянной:  
D(x) = 0, то х = М(х)
- Неравенство Коши-Буняковского: ковариация x и y не превышает корня из произведения их Дисперсий. Если равно, то x и y линейно зависимы. Или, другими словами, их коэффициент корреляции (cor) равен 1
- cor(x, y) = cov(x, y) / корень(D(x)D(y)). Если x и y н.с.в., то cor(x, y) = 0 т.к. cov = 0

## Схема независимых испытаний
Пусть мы некоторый эксперимент повторяем n раз. Независимость испытаний означает, что вероятность общего исхода равна произведению вероятностей элементарных исходов.   
Если n=2, то говорим о биномиальной схеме независимых испытаний или схемой Бернулли.

## Лекция 10: Математическая статистика - Финальное распределение в цепи Маркова <a name="Лекция10:Математическаястатистика-ФинальноераспределениевцепиМаркова"></a>

#### Продолжаем случайное блуждание вдоль оси
Ранее мы получили для симметричных блужданий:
- вероятность возврата в начальную точку за 2n шагов
- вероятность невозврата в начальную точку за 2n шагов
Закон арксинуса

#### Цепи Маркова
Матрица перехода за один шаг. Матрица перехода за m шагов, m=1,2,...
Случайные блуждания - это тоже цепь Маркова со счетным числом состояний.
Свойства матриц перехода:
1. элемент матрицы больше нуля, нщ меньше единицы
2. сумма элементов по строке равна единице
Матрица со свойствами 1 и 2 называется стохастической
3. Матрица перехода за два шага есть квадрат матрицы перехода за один шаг. Верно для любого количества шагов.
4. уравнение Чепмена-Колмогорова для дискретного пространства-времени

#### Поведение цепи Маркова при n стремящемся к бесконечности
Финальные вероятности цепи Маркова. Если они существуют, то возникает стационарность финального распределения.
В большом ансамбле систем среднее по времени равно среднему по ансамблю (свойство эргодичности)
